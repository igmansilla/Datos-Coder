```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

import seaborn as sns

url = 'https://raw.githubusercontent.com/igmansilla/Datos-Coder/main/df_limpio.csv'

df = pd.read_csv(url)
```

```python
df.head()
```

## Correlación de las variables

```python
variables = list(df.columns)[3:]
df[variables].corr()
```

```python
sns.lineplot('year', 'life_expectancy', data=df, marker='o')
plt.title('Esperanza de vida a traves de los años')
plt.show()
```

```python
df_developed = df[df.status == 'Developed']
```

```python
sns.lineplot('year', 'life_expectancy', data=df_developed, marker='o')
plt.title('Esperanza de vida a traves de los años. Paises desarrollados')
plt.show()
```

```python
df_developing = df[df.status == 'Developing']
```

```python
sns.lineplot('year', 'life_expectancy', data=df_developing, marker='o')
plt.title('Esperanza de vida a traves de los años. Paises en desarrollo')
plt.show()
```

Vemos que la esperanza de vida de los paises desarollados es mayor que la de los que todavía estan en desarrollo. Pero que sin embargo el crecimiento de los desarollados tuvo mas fluctuaciones e incluso una caída mas pronunciada
