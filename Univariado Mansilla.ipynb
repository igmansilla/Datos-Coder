```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

import seaborn as sns

url = 'https://raw.githubusercontent.com/igmansilla/Datos-Coder/main/df_limpio.csv'

df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.describe()
```

## Histogramas

```python
cont_vars = list(df.columns)[3:]
plt.figure(figsize=(15, 20))
for i, col in enumerate(cont_vars, 1):
    plt.subplot(5, 4, i)
    plt.hist(df[col])
    plt.title(col)
```

### Filas y paÃ­ses

```python
plt.figure(figsize=(15, 25))
df.country.value_counts(ascending=True).plot(kind='barh')
plt.title('Cantidad de filas por paÃ­s')
plt.xlabel('Cantidad de filas')
plt.ylabel('Pais')
plt.tight_layout()
plt.show()
```

El grÃ¡afico es desproporcionado pero muestra que hay algunos paises a los que les faltan filas

```python
df.year.value_counts().sort_index().plot(kind='barh')
plt.title('Cantidad de filas por aÃ±o')
plt.xlabel('Cantidad de filas')
plt.ylabel('AÃ±o')
plt.show()
```

El aÃ±o 2013 tiene algunos registros de mÃ¡s

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
df.status.value_counts().plot(kind='bar')
plt.title('Cantidad de filas por estado del pais')
plt.xlabel('Estado del paÃ­s')
plt.ylabel('Cantidad de filas')
plt.xticks(rotation=0)

plt.subplot(122)
df.status.value_counts().plot(kind='pie', autopct='%.2f')
plt.ylabel('')
plt.title('Estado del paÃ­s. GrÃ¡fico de torta')

plt.show()
```

La mayor parte de nuestros datos proceden de paÃ­ses clasificados como "en desarrollo"(developing): el 82,57% para ser exactos. Esto dificulta el uso de modelos predictivos porque hya un sesgo en la cantidad de datos de un valor y otro. Es probable que cualquier modelo que se utilice represente con mayor, ya que la mayorÃ­a de los datos se encuentran en paÃ­ses "en desarrollo" y no en paÃ­ses "desarrollados".
